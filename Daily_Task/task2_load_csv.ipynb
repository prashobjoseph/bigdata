```python
#importing libraries

import pandas as pd
import numpy as np
from sqlalchemy import create_engine
```

```python
file_path = "C:/Users/prash/Downloads/customers-1000-dataset.csv"
df = pd.read_csv(file_path)

print("Original Data:")

print(df.head())
```

```python
# Transformation 1: Rename columns to lowercase
df.columns = df.columns.str.upper()
print(df.head())
```

```python
# Transformation 2: Fill missing values in a specific column
df['COUNTRY'] = df['COUNTRY'].fillna(df['COUNTRY'].mode())
print(df.head(20))
```

```python
# Transformation 3: Create a new column based on existing data
df['FULLNAME'] = df['FIRST NAME'] + df['LAST NAME']
df['FULLNAME'].head()
```

```python
# Transformation 4:Add a new column 'Salary' with random values between 30000 and 100000
df['SALARY'] = np.random.randint(30000, 100000, size=len(df))
print(df.head(20))
```

```python
# Transformation 5: Filter rows where a condition is met
df = df[df['COUNTRY'] =="China"]  
print(df.head(7))
```

```python
#Connect to the Database
DATABASE_TYPE = 'mysql'  # 'mysql', 'sqlite', etc.
DB_DRIVER = 'pymysql'        # Database driver
DB_USER = 'root'
DB_PASSWORD = '7098'
DB_HOST = 'localhost'
DB_PORT = '3306'
DB_NAME = 'trial'

#Construct the database URL
DATABASE_URL = f"{DATABASE_TYPE}+{DB_DRIVER}://{DB_USER}:{DB_PASSWORD}@{DB_HOST}:{DB_PORT}/{DB_NAME}"

# Create a database connection
engine = create_engine(DATABASE_URL)

# Step 4: Load Transformed Data into the Database
table_name = "transformed_data"  # Desired table name in the database
df.to_sql(table_name, con=engine, if_exists='replace', index=False)  # Overwrites table if exists

print(f"\nData has been loaded into the table '{table_name}' in the database.")
```
